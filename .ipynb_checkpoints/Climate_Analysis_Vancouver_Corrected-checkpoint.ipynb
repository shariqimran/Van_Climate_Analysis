
# Comprehensive Climate Analysis: Vancouver
This notebook performs a detailed climate analysis for Vancouver, focusing on historical temperature and precipitation data. The project includes trend analysis, forecasting, and visualization to understand past changes and project future behavior.

- **Phase 2**: Exploratory Data Analysis (EDA)

## Phase 1: Data Loading and Preprocessing

```python

# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from prophet import Prophet

# Load dataset
data_path = '/Users/shariq/Desktop/Project/data/weatherstats_vancouver_normal_monthly.csv'
data = pd.read_csv(data_path)

# Inspect data
print("Dataset Overview:")
print(data.info())
print(data.head())

# Convert date column to datetime
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Clean and preprocess data
columns_to_drop = [col for col in data.columns if col.endswith('_c') or '_d' in col]
data = data.drop(columns=columns_to_drop)

# Create average temperature column
if 'max_temperature_v' in data.columns and 'min_temperature_v' in data.columns:
    data['avg_temperature_v'] = (data['max_temperature_v'] + data['min_temperature_v']) / 2
    data.drop(columns=['max_temperature_v', 'min_temperature_v'], inplace=True)

# Drop rows with missing critical data
cleaned_data = data.dropna(subset=['avg_temperature_v', 'precipitation_v']).set_index('date')

# Summary
print("Cleaned Data Summary:")
print(cleaned_data.describe())
```

## Phase 2: Exploratory Data Analysis (EDA)

```python

# Line plots for temperature and precipitation trends
plt.figure(figsize=(14, 7))
plt.plot(cleaned_data.index, cleaned_data['avg_temperature_v'], label='Average Temperature')
plt.plot(cleaned_data.index, cleaned_data['precipitation_v'], label='Precipitation')
plt.title('Trends in Temperature and Precipitation')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.show()

# Seasonal decomposition for temperature
decomposition = seasonal_decompose(cleaned_data['avg_temperature_v'], model='additive', period=12)
decomposition.plot()
plt.show()

# Correlation heatmap
sns.heatmap(cleaned_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Variables')
plt.show()
```

## Phase 3: Stationarity Testing and Transformation

```python

# Function for ADF test
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("Critical Values:", result[4])

# Perform ADF test on temperature
print("ADF Test for Average Temperature:")
adf_test(cleaned_data['avg_temperature_v'])

# Perform ADF test on precipitation
print("ADF Test for Precipitation:")
adf_test(cleaned_data['precipitation_v'])
```

## Phase 4: Modeling and Forecasting

```python

# Prepare data for Prophet
temperature_data = cleaned_data.reset_index()[['date', 'avg_temperature_v']].rename(columns={'date': 'ds', 'avg_temperature_v': 'y'})
precipitation_data = cleaned_data.reset_index()[['date', 'precipitation_v']].rename(columns={'date': 'ds', 'precipitation_v': 'y'})

# Prophet model for temperature
temp_model = Prophet()
temp_model.fit(temperature_data)
temp_future = temp_model.make_future_dataframe(periods=60, freq='M')
temp_forecast = temp_model.predict(temp_future)

# Plot temperature forecast
temp_model.plot(temp_forecast)
plt.title('Temperature Forecast with Prophet')
plt.show()

# Prophet model for precipitation
precip_model = Prophet()
precip_model.fit(precipitation_data)
precip_future = precip_model.make_future_dataframe(periods=60, freq='M')
precip_forecast = precip_model.predict(precip_future)

# Plot precipitation forecast
precip_model.plot(precip_forecast)
plt.title('Precipitation Forecast with Prophet')
plt.show()
```
