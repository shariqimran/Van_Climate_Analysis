
# Vancouver Climate Analysis Project

This project analyzes historical temperature and precipitation data for Vancouver. The goal is to explore trends, identify seasonal patterns, and forecast future values using advanced time-series modeling techniques.

### Objectives
1. Understand historical trends and seasonal behaviors in climate variables.
2. Train predictive models (Prophet) to forecast future temperature and precipitation.
3. Provide actionable insights through comprehensive visualizations and analysis.

---


## Phase 1: Data Loading and Preprocessing

```python

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.stattools import adfuller, kpss
from statsmodels.tsa.seasonal import seasonal_decompose
from prophet import Prophet

# Load the dataset
data_file_path = '/Users/shariq/Desktop/Project/data/weatherstats_vancouver_normal_monthly.csv'
data = pd.read_csv(data_file_path)

# Inspect the dataset
print("Dataset Information:")
print(data.info())
print(data.head())

# Convert 'date' column to datetime|
data['date'] = pd.to_datetime(data['date'], errors='coerce')

# Drop columns with irrelevant or auxiliary data
columns_to_drop = [col for col in data.columns if col.endswith('_c') or '_d' in col]
data.drop(columns=columns_to_drop, inplace=True)

# Create average temperature column if min/max columns exist
if 'max_temperature_v' in data.columns and 'min_temperature_v' in data.columns:
    data['avg_temperature_v'] = (data['max_temperature_v'] + data['min_temperature_v']) / 2
    data.drop(columns=['max_temperature_v', 'min_temperature_v'], inplace=True)

# Drop rows with missing values in critical columns
cleaned_data = data.dropna(subset=['avg_temperature_v', 'precipitation_v']).set_index('date')

# Summary statistics of cleaned data
print("Cleaned Data Summary:")
print(cleaned_data.describe())
```

## Phase 2: Exploratory Data Analysis (EDA)

```python

# Visualize trends in temperature and precipitation
plt.figure(figsize=(14, 7))
plt.plot(cleaned_data.index, cleaned_data['avg_temperature_v'], label='Average Temperature')
plt.plot(cleaned_data.index, cleaned_data['precipitation_v'], label='Precipitation')
plt.title('Trends in Temperature and Precipitation')
plt.xlabel('Year')
plt.ylabel('Values')
plt.legend()
plt.show()

# Seasonal decomposition for temperature
decomposition = seasonal_decompose(cleaned_data['avg_temperature_v'], model='additive', period=12)
decomposition.plot()
plt.show()

# Correlation heatmap
sns.heatmap(cleaned_data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Climate Variables')
plt.show()
```

## Phase 3: Stationarity Testing and Transformation

```python

# ADF Test function
def adf_test(series):
    result = adfuller(series, autolag='AIC')
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("Critical Values:", result[4])
    return result[1]  # Return p-value

# Check stationarity for temperature and precipitation
print("ADF Test for Average Temperature:")
adf_temp = adf_test(cleaned_data['avg_temperature_v'])

print("ADF Test for Precipitation:")
adf_precip = adf_test(cleaned_data['precipitation_v'])

# Apply transformations if necessary
if adf_temp > 0.05:
    print("Temperature series is non-stationary. Applying first differencing...")
    cleaned_data['avg_temperature_v_diff'] = cleaned_data['avg_temperature_v'].diff().dropna()
if adf_precip > 0.05:
    print("Precipitation series is non-stationary. Applying first differencing...")
    cleaned_data['precipitation_v_diff'] = cleaned_data['precipitation_v'].diff().dropna()
```

## Phase 4: Modeling and Forecasting with Prophet

```python

# Prepare data for Prophet
temperature_data = cleaned_data.reset_index()[['date', 'avg_temperature_v']].rename(columns={'date': 'ds', 'avg_temperature_v': 'y'})
precipitation_data = cleaned_data.reset_index()[['date', 'precipitation_v']].rename(columns={'date': 'ds', 'precipitation_v': 'y'})

# Prophet model for temperature
temp_model = Prophet()
temp_model.add_country_holidays(country_name='CA')  # Add Canadian holidays for seasonality
temp_model.fit(temperature_data)
temp_future = temp_model.make_future_dataframe(periods=60, freq='M')
temp_forecast = temp_model.predict(temp_future)

# Plot temperature forecast
temp_model.plot(temp_forecast)
plt.title('Temperature Forecast with Prophet')
plt.show()

# Prophet model for precipitation
precip_model = Prophet()
precip_model.add_country_holidays(country_name='CA')  # Add Canadian holidays for seasonality
precip_model.fit(precipitation_data
precip_future = precip_model.make_future_dataframe(periods=60, freq='M')
precip_forecast = precip_model.predict(precip_future)

# Plot precipitation forecast
precip_model.plot(precip_forecast)
plt.title('Precipitation Forecast with Prophet')
plt.show()
```

```python

# Prepare data for Prophet
temperature_data = cleaned_data.reset_index()[['date', 'avg_temperature_v']].rename(columns={'date': 'ds', 'avg_temperature_v': 'y'})

# Initialize Prophet with additional settings
temp_model = Prophet(yearly_seasonality=True, changepoint_prior_scale=0.1)
temp_model.add_country_holidays(country_name='CA')  # Canadian holidays

# Fit the model
temp_model.fit(temperature_data)

# Make future dataframe and predict
temp_future = temp_model.make_future_dataframe(periods=60, freq='M')
temp_forecast = temp_model.predict(temp_future)

# Plot forecast
fig1 = temp_model.plot(temp_forecast)
plt.title('Advanced Temperature Forecast with Prophet')
plt.show()

# Analyze components: Trend, Seasonality, and Holidays
fig2 = temp_model.plot_components(temp_forecast)
plt.show()

# Detect and visualize changepoints
from prophet.plot import add_changepoints_to_plot
fig3 = add_changepoints_to_plot(fig1.gca(), temp_model, temp_forecast)
plt.title('Changepoints in Temperature Trends')
plt.show()
```

```python

# Add rolling averages and other predictors as regressors
temperature_data['rolling_precip'] = cleaned_data['precip_rolling'].values

# Initialize Prophet with additional regressors
multi_var_model = Prophet()
multi_var_model.add_regressor('rolling_precip')

# Fit the model
multi_var_model.fit(temperature_data)

# Forecast with regressors
multi_var_future = multi_var_model.make_future_dataframe(periods=60, freq='M')
multi_var_future['rolling_precip'] = cleaned_data['precip_rolling'].iloc[-60:].values  # Use latest data for predictions
multi_var_forecast = multi_var_model.predict(multi_var_future)

# Plot forecast
multi_var_model.plot(multi_var_forecast)
plt.title('Multi-Variable Temperature Forecast with Precipitation Regressor')
plt.show()
```
